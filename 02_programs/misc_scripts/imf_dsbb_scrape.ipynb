```python
# Import libraries
# Python program to scrape table from website
  
# import libraries selenium and time
from selenium import webdriver
from time import sleep
import pandas as pd
from bs4 import BeautifulSoup
import re
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize 
import os

from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys

# Setup chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Ensure GUI is off
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

os.environ['WDM_SSL_VERIFY'] = '0'

path = "C:/Users/WB469649/WBG/Statistical Performance Indicators (SPI) - WB Group - Documents/Data/SPI 2024/01_raw_data/"

```

```python
#read in data for Employment
empl_df=pd.read_csv(r'%s/D5.2.6.EMPL.2023.csv' % (path))
empl_df=empl_df.rename(columns={'IMF_checker_EMP':'IMF_checker_EMP_2023'})
empl_df=empl_df.rename(columns={'IMF_checker_UNEMP':'IMF_checker_UNEMP_2023'})

cnin_df=pd.read_csv(r'%s/D5.2.3.CNIN.2023.csv' % (path))


fina_df=pd.read_csv(r'%s/D5.2.8.FINA.2023.csv' % (path))
fina_df=fina_df.rename(columns={'FINA':'FINA_Status_2023'})
```

```python
# Set path to chromedriver as per your configuration
webdriver_service = Service(ChromeDriverManager().install())

# Choose Chrome Browser
driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)

#function to check if string is in text
def text_checker(text,data):
    flag=data.Text.apply(lambda x: bool(re.search(text,x)))
    t=flag.max()
    return t

def imf_get_contents(iso):

    url = "https://dsbb.imf.org/egdds/dqaf-base/country/%s/category/EMP00" % (iso)
    
    # Choose Chrome Browser
    driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)
    # Get the website
    driver.get(
        url)
    
    # Make Python sleep for some time
    sleep(3)

    html=driver.page_source
    try:
        # No error
        table = pd.read_html(html)
        df = table[0]
        df=df.iloc[:, [0,1]]
        df.columns=['Type', 'Text']
        if len(df)<1:
            return "No Data"
        else:    
            #keep just the section on classifications
            df = df.loc[df.Type.astype(str).str.contains('^2.*'), :]
            df=df.reset_index(drop=True)
            driver.quit()
            if len(df)<1:
                return "No Data"
            elif text_checker("NAICS", df):
                return "NAICS"
            elif text_checker("ISCE",df):
                return "ICSE"    
            elif text_checker("International Standard Classification of Status in Employment",df):
                return "International Standard Classification of Status in Employment"         
            elif text_checker("International Labour Organisation",df):
                return "International Labour Organisation"      
            elif text_checker("\\b(ILO)\\b",df):
                return "ILO"      
            else:
                return "Not Found"
    except:
        # Error
        return "Error" 

        


def imf_get_contents_ur(iso):

    url = "https://dsbb.imf.org/egdds/dqaf-base/country/%s/category/UEM00" % (iso)
    
    # Choose Chrome Browser
    driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)
    # Get the website
    driver.get(
        url)
    
    # Make Python sleep for some time
    sleep(3)
    html=driver.page_source
    try:
        # No error
        table = pd.read_html(html)
        df = table[0]
        df=df.iloc[:, [0,1]]
        df.columns=['Type', 'Text']
        if len(df)<1:
            return "No Data"
        else:    
            #keep just the section on classifications
            df = df.loc[df.Type.astype(str).str.contains('^2.*'), :]
            df=df.reset_index(drop=True)
            driver.quit()
            if len(df)<1:
                return "No Data"
            elif text_checker("NAICS", df):
                return "NAICS"
            elif text_checker("ISCE",df):
                return "ICSE"    
            elif text_checker("International Standard Classification of Status in Employment",df):
                return "International Standard Classification of Status in Employment"         
            elif text_checker("International Labour Organisation",df):
                return "International Labour Organisation"      
            elif text_checker("\\b(ILO)\\b",df):
                return "ILO"                         
            else:
                return "Not Found"
    except:
        # Error
        return "Error" 





def imf_get_contents_cnin(iso):

    url = "https://dsbb.imf.org/egdds/dqaf-base/country/%s/category/IND00" % (iso)
 
    
    # Choose Chrome Browser
    driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)
        # Get the website
    driver.get(
        url)
    
    # Make Python sleep for some time
    sleep(3)

    html=driver.page_source
    try:
        # No error
        table = pd.read_html(html)
        df = table[0]
        df=df.iloc[:, [0,1]]
        df.columns=['Type', 'Text']
        if len(df)<1:
            return "No Data"
        else:    
            #keep just the section on classifications
            df = df.loc[df.Type.astype(str).str.contains('^2.*'), :]
            df=df.reset_index(drop=True)
            driver.quit()
            if len(df)<1:
                return "No Data"
            elif text_checker("rev 3", df):
                return "rev3"
            elif text_checker("rev 4",df):
                return "rev4"    
            elif text_checker("NACE",df):
                return "nace"                    
            elif text_checker("International Recommendations for the Index of Industrial Production",df):
                return "rev4"           
            else:
                return "Not Found"
    except:
        # Error
        return "Error" 

def imf_get_contents_fina(iso):

    url = "https://dsbb.imf.org/egdds/dqaf-base/country/%s/category/CGO00" % (iso)
    
    # Choose Chrome Browser
    driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)
    # Get the website
    driver.get(
        url)
    
    # Make Python sleep for some time
    sleep(3)

    html=driver.page_source
    try:
        # No error
        table = pd.read_html(html)
        df = table[0]
        df=df.iloc[:, [0,1]]
        df.columns=['Type', 'Text']
        if len(df)<1:
            return "No Data"
        else:    
            #keep just the section on classifications
            df = df.loc[df.Type.astype(str).str.contains('^2.*'), :]
            df=df.reset_index(drop=True)
            driver.quit()
            if len(df)<1:
                return "No Data"
            elif text_checker("Government Finance Statistics Manual 2014", df):
                return "2014"
            elif text_checker("Government Finance Statistics Manual 1986",df):
                return "1986"    
            elif text_checker("Government Finance Statistics Manual 2001",df):
                return "2001"   
            elif text_checker("GFSM 1986",df):
                return "1986" 
            elif text_checker("GFSM 2001",df):
                return "2001"                           
            elif text_checker("2014",df):
                return ' '.join(df['Text'])    
            elif text_checker("1986",df):
                return ' '.join(df['Text'])    
            elif text_checker("2001",df):
                return ' '.join(df['Text'])      
            elif text_checker("Government Finance Statistics Manual",df):
                return ' '.join(df['Text'])       
            elif text_checker("GFSM",df):
                return ' '.join(df['Text'])                      
            else:
                return "Not Found"
    except:
        # Error
        return "Error" 
```

```python
#read in file from the ILO and use as a basis for EMPL classifications as well.  Check if  ISCE mentioned in classifications fields for labor force surveys

ilo_surveys=pd.read_csv('C:/Users/WB469649/WBG/Statistical Performance Indicators (SPI) - WB Group - Documents/Data/SPI 2024/01_raw_data/ilo_sources_catalog.csv')

#Keep just the labor force surveys
ilo_surveys=ilo_surveys.loc[ilo_surveys['Source type']=='Labour force survey',:]

#read in ILO country names and codes file
ilo_codes=pd.read_csv(r'%s/ILO_reference_area.csv' % (path))   
#rename column ref_area.label to Country
ilo_codes=ilo_codes.rename(columns={' ref_area.label':'Country',
                                    'ref_area':'iso3c'})    
        
#merge the two files
ilo_surveys=pd.merge(ilo_surveys, ilo_codes, on='Country', how='left')

#create a new column checking if ISCE mentioned in international classification column
ilo_surveys['ISCE']=ilo_surveys['International classifications'].apply(lambda x: bool(re.search("ICSE",str(x)))).astype(int)

#grouping by iso3c find the max value of ISCE
ilo_surveys=ilo_surveys.groupby('iso3c').agg({'ISCE':'max'}).reset_index()

#MERGE ILO SURVEYS WITH EMPL_DF
#empl_df.rename(columns={'Code':'iso3c'}, inplace=True)
empl_df_complete=pd.merge(empl_df, ilo_surveys, on='iso3c', how='left')

empl_df_complete.to_csv("C:/Users/WB469649/WBG/Statistical Performance Indicators (SPI) - WB Group - Documents/Data/SPI 2024/03_output/D1.6.MSC.EMPL.csv")
```
